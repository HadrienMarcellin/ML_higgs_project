```python
import numpy as np
import math
import csv
from proj1_helpers import *
import matplotlib.pyplot as plt
```

```python
y, tx, ids = load_csv_data("data_csv/train.csv", sub_sample=True)
```

```python
print("y has shape {0}".format(y.shape))
print("X has shape {0}".format(tx.shape))
```

### Feature Distribution

```python
n_bins = 20
fig, axs = plt.subplots(15, 2, sharey=False, squeeze=False, tight_layout=False, figsize=(15, 60))

for i, feature in enumerate(tx.T):
    col = i % 2
    row = int((i-col)/2)
    feature_true, feature_false, nb_nan = split_data_according_to_truth(feature, y)
    axs[row, col].hist([feature_true, feature_false], histtype = 'bar',bins=n_bins, label=['True', 'False'])
    axs[row, col].set_title("Feature n {0} with {1} NAN".format(i, nb_nan))
    axs[row, col].legend(prop={'size': 10})

plt.show()
```

```python
def split_data_according_to_truth(x,y):
    '''Attention, tx et y doivent tre ranger dans le mme ordre'''
    x_true = x[np.where((y>0) & (x > -900))]
    x_false = x[np.where((y<0) & (x > -900))]
    nb_nan = len(x[np.where(x<-900)])
    return x_true, x_false, nb_nan
```

#### Missing Data

```python
temp = tx.copy()
temp[np.where(temp>-800)] = 0
temp[np.where(temp<-800)] = 1
nan_sum_per_feature = temp.sum(axis=0)
nan_sum_per_sample = temp.sum(axis=1)
```

```python
fig, axs = plt.subplots(1, 2, sharey=False, squeeze=True, tight_layout=False, figsize=(10, 5))
axs[0].hist(nan_sum_per_sample, bins=20, histtype='bar')
axs[0].set_xlabel("number of NAN values")
axs[0].set_ylabel("number of samples")

axs[1].hist(nan_sum_per_feature, bins=20, histtype='bar')
axs[1].set_xlabel("number of NAN values")
axs[1].set_ylabel("number of features")

plt.show()
```
